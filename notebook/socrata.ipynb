4. Make an API request that returns the entire Top 500 Monthly Searches dataset. Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.

```python
query3 = endpoint + "?" + "$query=SELECT year, SUM(query_count) WHERE query_text LIKE 'maps' GROUP BY year"
var3 = requests.get(query3).json()
var3
```

```python
df3 = pd.DataFrame(var3)
df3
```

```python
print(df3.columns)
```

```python
df3['SUM_query_count'] = df3['SUM_query_count'].astype(int)
df3
```

```python
years = df3['year']
counts = df3['SUM_query_count']

plt.plot(years, counts)
plt.xlabel("Month")
plt.ylabel("Number of searches for 'maps'")
plt.title("Monthly searches for 'maps' (Top 500)")
plt.show()
```
